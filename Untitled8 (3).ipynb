```python
print(all_data.isnull().values.any())
print(all_data.isnull().values)
print(all_data.isnull().any().any())
```

```python
all_data['Posted On'] = pd.to_datetime(all_data["Posted On"])
```

```python
def one_hot_encoding(df,col):
    oneH = pd.get_dummies(df[col])
    df = df.drop(col,axis = 1)
    df = df.join(oneH)
    return df
```

```python
columns = ['Area Type','City','Furnishing Status','Tenant Preferred','Point of Contact']
for col in columns:
    all_data = one_hot_encoding(all_data, col)
```

```python
all_data = all_data.join(all_data['Floor'].str.split(' out of ', 1, expand=True).rename(columns={0:'Floor Level', 1:'Floor Max'}))
all_data['Floor Level'] = all_data.apply(lambda x: 0 if x['Floor Level'] =='Ground' \
                               else ( -1 if x['Floor Level'] =='Lower Basement' else (x['Floor Max']) ) , axis=1)
```

```python
all_data = all_data.dropna()
all_data['Floor Level'] = all_data['Floor Level'].astype(int)
all_data['Floor Max'] = all_data['Floor Max'].astype(int)
```

```python
# DATA VALUES 
all_data['month posted'] = all_data['Posted On'].dt.month
all_data['day posted'] = all_data['Posted On'].dt.day
all_data['day of week posted'] = all_data['Posted On'].dt.day_of_week
all_data['quarter poster'] = all_data['Posted On'].dt.quarter
```

```python
all_data = all_data.drop(['Floor','Posted On','Area Locality'], axis=1)
all_data = all_data.dropna()
```

```python
# DUPLICATES

print(all_data.duplicated().sum())
all_data = all_data.drop_duplicates()
```

```python
all_data['Rent'] = np.log1p(all_data['Rent'])
```

```python
target = 'Rent'
X = all_data.loc[:,all_data.columns!=target]
y = all_data.loc[:,all_data.columns==target]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.20, shuffle=True)
x_train
```

```python
x_test
```

```python
y_train
```

```python
print("x_train shape: ", x_train.shape)
print("x_test shape: ", x_test.shape)
print("y_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#build and train a linear regression model
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
plt.scatter(lr.predict(x_train), y_train)
plt.xlabel('Predicted value of Y')
plt.ylabel('Real value of Y')
plt.show()
```

```python
#linear regression
import sklearn.metrics as metrics
lr_prediction = lr.predict(x_test)
mae_lr = metrics.mean_absolute_error(y_test, lr_prediction)
mse_lr =  metrics.mean_squared_error(y_test, lr_prediction)
rmse_lr =  np.sqrt(mse_lr)
print('MAE:', mae_lr)
print('MSE:', mse_lr)
print('RMSE:', rmse_lr)
model=lr.fit(X,y)
print('Model coeff : ',model.coef_)
print('Model intercept : ',model.intercept_)
print('Accuracy : ', model.score(X,y))
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
parameter={'alpha':[0.5,1,2,3,4,5,10,20,30,40,50]}
ridgecv=GridSearchCV(ridge,parameter,scoring='neg_mean_squared_error',cv=5)
ridgecv.fit(x_train,y_train)
```

```python
print(ridgecv.best_params_)
ridge_pred=ridgecv.predict(x_test)
sns.displot(ridge_pred,kind='kde')
sns.displot(y_test,kind='kde')
```

```python
from sklearn.metrics import  r2_score 
rscore=r2_score(ridge_pred,y_test)
rscore
```

```python
 import sklearn.metrics as metrics
ridge_prediction = ridgecv.predict(x_test)
mae_ridge = metrics.mean_absolute_error(y_test, ridge_prediction)
mse_ridge =  metrics.mean_squared_error(y_test, ridge_prediction)
rmse_ridge=  np.sqrt(mse_ridge)
print('MAE:', mae_ridge)
print('MSE:', mse_ridge)
print('RMSE:', rmse_ridge)
model=ridge.fit(X,y)
print('Model coeff : ',model.coef_)
print('Model intercept : ',model.intercept_)
print('Accuracy : ', model.score(X,y))
```

```python
#Lasso Regreesion
from sklearn.linear_model import Lasso
lasso=Lasso()
parameter={'alpha':[0.01,0.001,0.0001,0.5,1]}
lassocv=GridSearchCV(lasso,parameter,scoring='neg_mean_squared_error',cv=5)
lassocv.fit(x_train,y_train)
```

```python
print(lassocv.best_params_)
lasso_pred=lassocv.predict(x_test)
lsscore=r2_score(lasso_pred,y_test)
lsscore
```

```python
 import sklearn.metrics as metrics
lasso_prediction = lassocv.predict(x_test)
mae_lasso = metrics.mean_absolute_error(y_test, lasso_prediction)
mse_lasso =  metrics.mean_squared_error(y_test, lasso_prediction)
rmse_lasso=  np.sqrt(mse_lasso)
print('MAE:', mae_lasso)
print('MSE:', mse_lasso)
print('RMSE:', rmse_lasso)
model=lasso.fit(X,y)
print('Model coeff : ',model.coef_)
print('Model intercept : ',model.intercept_)
print('Accuracy : ', model.score(X,y))
```
